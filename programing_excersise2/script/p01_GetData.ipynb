```python
def read_fasta_file(data_dir):
    import math
    from Bio import SeqIO
    import matplotlib.pyplot as plt
    import numpy as np
    import os

    pathes = glob.glob(f"{data_dir}/*")
    for path in pathes:
        sequences = []
        for record in SeqIO.parse(path, "fasta"):
            sequences.append(record.seq)
        sequences.append(sequences)

        length = []
        for sequences in sequences:
            for sequence in sequences:
                length.append(len(sequence))
        length.append(length)
    return sequence, length

data_dir = "../data"
pathes = glob.glob(data_dir)
sequences, lengthes = read_fasta_file(data_dir)
```

```python



for i in range(len(pathes))[:2]:
    path = pathes[i]
    species = os.path.basename(path).split(".")[0]
    genome = sequences[i]

    if not os.path.exists(f"../output/compossision/"):
        os.mkdir(f"../output/compossision/")
    if not os.path.exists(f"../output/compossision/{species}"):
        os.mkdir(f"../output/compossision/{species}")

    n_chr = len(genome)
    genome_df_1, genome_df_2, genome_df_3 = DataFrame(), DataFrame(), DataFrame()
    for j, chr in enumerate(genome):
        chr_array = str(chr.lower())
        
        base_dict_1, base_dict_2, base_dict_3 = defaultdict(int), defaultdict(int), defaultdict(int)

        for k in range(len(chr_array)):
            base_dict_1[chr_array[k:k+1]] += 1
            if k<len(chr_array)-3:
                base_dict_3[chr_array[k:k+3]] += 1
                if k<len(chr_array)-2:
                    base_dict_2[chr_array[k:k+2]] += 1
            chr_df_1 = DataFrame(data=base_dict_1.values(), index=base_dict_1.keys(), columns=[f"chr_{j}"])
            chr_df_2 = DataFrame(data=base_dict_2.values(), index=base_dict_2.keys(), columns=[f"chr_{j}"])
            chr_df_3 = DataFrame(data=base_dict_3.values(), index=base_dict_3.keys(), columns=[f"chr_{j}"])

        genome_df_1 = pd.concat([genome_df_1, chr_df_1], axis=1)
        genome_df_2 = pd.concat([genome_df_2, chr_df_2], axis=1)
        genome_df_3 = pd.concat([genome_df_3, chr_df_3], axis=1)
        print(f"chr_{j} done")
        

    print(species)
    for j, df in enumerate([genome_df_1, genome_df_2, genome_df_3]):
        df.to_csv(f"../output/compossision/{species}/length_{j+1}.csv")    
        
        



        



```
